```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#functional method
plt.plot(x, y, 'r-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x vs y')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r-')

plt.subplot(1,2,2)
plt.plot(y,x, 'b-')
```

```python
#object oriented
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes.plot(x,y)
axes.set_xlabel('x lable')
axes.set_ylabel('y label')
axes.set_title("hello")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title("bada dabba")
axes2.plot(y,x)
axes2.set_title("chhota dabba")
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
# axes.plot(x,y)
# plt.tight_layout()
# for current_ax in axes:
#     current_ax.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title("first")
axes[1].plot(y,x)
axes[1].set_title("second")
```

```python
#figure size and DPI
```

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(3,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label= "x squanred")
ax.plot(x, x**3, label = "x cubed")

ax.legend(loc = (0.1, 0.1))
```

## plot appearance


```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = "pink", linewidth = 3)
ax.plot(x, x**3, color = "red", linewidth = 20, alpha = 0.5)
ax.plot(x, x**4, color = "blue", linewidth = 10, linestyle = "--")
```
